```python
import pandas as pd
import numpy as np
import matplotlib as plt
import datetime
import csv

import folium
from folium.features import GeoJson, GeoJsonTooltip, GeoJsonPopup

import geopandas as gpd
from functools import partial

# 43°34′49″N 16°35′49″E
```

```python
# 1 Dugopolje mapa

# Координаты центра Dugopolje
dugopolje_coords = [43.5867, 16.5833] # [43.6667, 30.5833] [52.5200, 13.4050]

#import folium
#m = folium.Map(location=[latitude, longitude], zoom_start=10)
#folium.Marker([lat, lon], popup='Marker Name').add_to(m)
#m.save('map.html')

dugopolje_map = folium.Map(location=dugopolje_coords,                        
                        tiles='OpenStreetMap',                        
                        scrollWheelZoom=False,                                               
                        crs='EPSG3857',
                        control_scale=True,                        
                        zoom_control=True,
                        dragging=True,
                        min_zoom=10,
                        max_zoom=25,
                        zoom_start=12)


# location=None, popup=None, tooltip=None, icon=None, draggable=False, **kwargs)
folium.Marker(location=dugopolje_coords,
              popup='Antalja',
              draggable=False,
              ).add_to(dugopolje_map)

dugopolje_map
```

```python
# 2 Data from today's monitoring stations:
# https://aqicn.org/data-platform/

dfp_today = pd.read_csv('data_Dugopolje.csv') 
dfp_today.head()
today = datetime.date.today()
dfd_today = dfp_today[dfp_today['date']== str(today)]
print(f'active {len(dfd_today)} counters:')
dfd_today 
```

```python
# 3 Air Pollution pm25 - анализ показателей pm25  dugopolje

# data
# dfd_today.index #['station_id']
# print(dfd_today.index, end='' )

#dfd_today['station_id']

print(f'{dfd_today.index}')
print(f'counters: ')
for x in dfd_today.index: #dfp_today['station_id']:
    print( dfd_today['station_id'][x], end=',')     
print()   

values = dfd_today['pm25']
mmin = min(values)
mmax = max(values)
delta= (mmax - mmin) 
print(f'{mmax=} {mmin=} {delta=} ')    

folium.CircleMarker( location=[43.5106, 16.4493 ], 
                         radius=6,  # Радиус круга в пикселях
                         popup=str(x),
                         color='green' ,  # Цвет круга
                         fill=True,
                         fill_color='green' 
                         ).add_to(dugopolje_map)

for x in dfd_today.index:      
    print( x, dfd_today['pm25'][x], dfd_today['lat'][x], dfd_today['long'][x] )

    if  mmin <= int( dfd_today['pm25'][x] ) <= mmin+delta/2:
        #print('*')
        c = 'g' #'#DA70D6'
        folium.CircleMarker( location=[dfd_today['lat'][x], dfd_today['long'][x] ], 
                         radius=6,  # Радиус круга в пикселях
                         popup=str(x),
                         color='green' ,  # Цвет круга
                         fill=True,
                         fill_color='green' 
                         ).add_to(dugopolje_map)
    elif  mmin+delta/2 < int( dfa_today['o3'][x] ) <= mmax:
        c = 'r' # '#8A2BE2'        
        folium.CircleMarker( location=[dfd_today['lat'][x], dfd_today['long'][x] ], 
                         radius=6,  # Радиус круга в пикселях
                         popup=str(x),
                         color='red' ,  # Цвет круга
                         fill=True,
                         fill_color='red'
                         ).add_to(dugopolje_map)


print(today)
print(f'Air Pollution pm25: {mmin=} {mmax=} {delta=}')     

dugopolje_map
```
