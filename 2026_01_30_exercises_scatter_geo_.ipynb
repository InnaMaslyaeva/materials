```python
# 2d eddition
# 1 Coordinates
import requests
# install plotly
import plotly.express as px

# 1 ISS coordinates
url = "http://api.open-notify.org/iss-now.json"
response = requests.get(url)
data = response.json()

pos = data["iss_position"]
lat = float(pos["latitude"])
lon = float(pos["longitude"])
#print(f'coordinates: {lat= } {lon= }')

# 2 Plotting
# https://plotly.com/python-api-reference/generated/plotly.express.scatter_geo.html
title = f"ISS Live Position: {lat= } {lon= } "
fig = px.scatter_geo(
    lat=[lat], lon=[lon],
    size_max=2,
    title=title,
    width=900, height=600,
    projection="natural earth"
)

fig.update_layout(
    autosize=True,
    #autosize=dict(
    #    type='fit',          # 'fit', 'contain', 'fill'
    #    fill=True,           # заполнять контейнер
    #    contain=True ),      # помещаться в контейнер

    #margin=dict(autoexpand=True),
    margin=dict(l=10, r=10, t=10, b=10, pad=5, ),
    paper_bgcolor='lightgray',
    title=dict(
       y=0.96,  # поднимаем заголовок еще выше
       x=0.3,
       xanchor='center',
       yanchor='top',
       font=dict(size=20)),
    geo=dict(
        #center=dict(lat=lat, lon=lon),
        projection_scale=1,  # регулируйте масштаб
        showland=True,
        landcolor="lightgray",
        showocean=True,
        oceancolor="lightblue",
        showcountries=True,
        countrycolor="darkgrey" ,   )
)

# fig.update_layout(
#     #autosize=True,  # автоматическая подгонка
#     margin=dict(l=100, r=0, t=0, b=0),
#     #title=dict(
#     #    y=0.1,  # поднимаем заголовок еще выше
#     #    x=0.1,
#     #    xanchor='center',
#     #    yanchor='top'),
#     geo=dict(
#         center=dict(lat=lat, lon=lon),
#         projection_scale=1,  # регулируйте масштаб
#         showland=True,
#         landcolor="lightgray",
#         showocean=True,
#         oceancolor="lightblue",
#         showcountries=True,
#         countrycolor="darkgrey"   )
# )

# Mаркер покрупнее
fig.update_traces(marker=dict(size=15))

fig.show()

# Посмотреть текущий layout
#print(fig.layout)

# Посмотреть текущий margin
#print(fig.layout.margin)

# Cловарь с данными
#rint(fig.to_dict()['layout'])
```

```python
# 3 Data collection
import requests
import time

# 1 ISS coordinates
res = []
for n in 1,2,3:
  url = "http://api.open-notify.org/iss-now.json"
  response = requests.get(url)
  data = response.json()
  pos = data["iss_position"]
  lat = float(pos["latitude"])
  lon = float(pos["longitude"])
  print(f'coordinates: {lat= } {lon= }')
  res.append([lat, lon])
  time.sleep(30)
print(f'{res=} ')
```
