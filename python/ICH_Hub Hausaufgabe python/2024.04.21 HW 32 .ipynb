```python
# 2024.04.21 HW 32
'''1. Реализовать класс Counter, который представляет счетчик. Класс должен поддерживать следующие операции:
- Увеличение значения счетчика на заданное число (оператор +=)
- Уменьшение значения счетчика на заданное число (оператор -=)
- Преобразование счетчика в строку (метод __str__)
- Получение текущего значения счетчика (метод __int__)
Пример использования:
counter = Counter(5)
counter += 3
print(counter)  # Вывод: 8
counter -= 2
print(int(counter))  # Вывод: 6'''

class Counter:
    def __init__(self, number):
        self.number = number

    def __add__(self, other):  
    # Увеличение значения счетчика на заданное число (оператор +=)    
        return self.number + other
    
    def __sub__(self, other):  
    # Уменьшение значения счетчика на заданное число (оператор -=)   
        return self.number - other

    def __str__(self):   
        # Преобразование счетчика в строку (метод __str__)     
        return str(self.number)
        
    def __int__(self):
        # Получение текущего значения счетчика (метод __int__)
        return self.number        

n = 5
counter = Counter(n)
print(f' {counter.number= } ')

counter += 1
print(">>>", counter)
counter -= 1
print(">>>", counter)

print(f' {counter.__str__()= } ')
print(f' {counter.__int__()= } ')
```

```python
''' 2. Реализовать класс Email, представляющий электронное письмо. Класс должен поддерживать следующие операции:
- Сравнение писем по дате (операторы <, >, <=, >=, ==, !=)
- Преобразование письма в строку (метод __str__)
- Получение длины текста письма (метод __len__)
- Получение хеш-значения письма (метод __hash__)
- Проверка наличия текста в письме (метод __bool__)
Пример использования:
email1 = Email("john@example.com", "jane@example.com", "Meeting", "Hi Jane, let's have a meeting tomorrow.", "2022-05-10")
email2 = Email("jane@example.com", "john@example.com", "Re: Meeting", "Sure, let's meet at 2 PM.", "2022-05-10")
email3 = Email("alice@example.com", "bob@example.com", "Hello", "Hi Bob, how are you?", "2022-05-09")
print(email1)  # Вывод:
"""
From: john@example.com
To: jane@example.com
Subject: Meeting
Hi Jane, let's have a meeting tomorrow.
"""
print(len(email2))  # Вывод: 24
print(hash(email3))  # Вывод: -920444056
print(bool(email1))  # Вывод: True
print(email2 > email3)  # Вывод: True '''

#import time
from functools import total_ordering

@total_ordering
class Email:
    def __init__(self, adressTo, addressFrom, Teme, Text, Time):
        self.adressTo = adressTo
        self.addressFrom = addressFrom
        self.Teme = Teme
        self.Text = Text
        self.Time = Time
    
    def __str__(self):
        #return f' {email.adressTo}  {email.addressFrom}  {email.Teme}  {email.Text}  {email.Time}' 
        return f'From: {email.adressTo}\nTo: {email.addressFrom}\nSubject: {email.Teme}\n{email.Text}'  
    
    def __len__(self):
        # Получение длины текста письма (метод __len__)
        return len( self.Text )
         
    def __hash__(self):
        # Получение хеш-значения письма (метод __hash__)
        # (self.adressTo, self.addressFrom, self.Teme, self.Text, self.Time)
        return hash( self.Text ) 
    
    def __bool__(self):
        # Проверка наличия текста в письме (метод __bool__)
        return len( self.Text ) > 0
    
    def __eq__(self, other):  
        # Сравнение писем по дате (операторы <, >, <=, >=, ==, !=)         
        return self.Time == other.Time
    
    def __lt__(self, other):
        # Сравнение писем по дате (операторы <, >, <=, >=, ==, !=)
        return self.Time < other.Time
    
email1 = Email("john@example.com", "jane@example.com", "Meeting", "Hi Jane, let's have a meeting tomorrow.", "2022-05-10")
email2 = Email("jane@example.com", "john@example.com", "Re: Meeting", "Sure, let's meet at 2 PM.", "2022-05-10")
email3 = Email("alice@example.com", "bob@example.com", "Hello", "Hi Bob, how are you?", "2022-05-09")

for email in email1, email2, email3:
    print( f'{email }' )    
    print( f'{len(email)= }' ) # = print(email.__len__())
    print( f'{hash(email)= }' )
    print( f'{bool(email)= }\n' )

print( 'Emails time analyses ')   
print( f'{email1.Time= } {email2.Time= } {email1 < email2= } {email1 == email2= } {email1 < email2= }'  )
print( f'{email1.Time= } {email3.Time= } {email1 < email3= } {email1 == email3= } {email1 < email3= }'  )   
print( f'{email2.Time= } {email3.Time= } {email2 < email3= } {email2 == email3= } {email2 < email3= }'  )  
```
