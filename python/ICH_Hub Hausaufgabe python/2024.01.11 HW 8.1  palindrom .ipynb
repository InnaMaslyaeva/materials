```python
# 2024.01.11 HW 8 
'''1. Напишите программу, которая запрашивает у пользователя 
целое число и определяет, является ли оно палиндромом. 
Число является палиндромом, если оно
читается одинаково слева направо и справа налево. 
Выведите соответствующее сообщение на экран с помощью команды print. 
Используйте математические операции. 
Работу со строками использовать нельзя.
Пример вывода: Введите целое число: 12321
Число является палиндромом.'''

x1 = x0 = x = int(input('Введите целое число: '))

l = 0 
while x0 > 0:
    l = l + 1
    x0 = x0 // 10

y = 0
while x1 > 0:
    n = x1 % 10
    y = y * 10 + n
    x1 = x1 // 10    
     
if x == y: print('Число является палиндромом.')
else: print('Число не является палиндромом.')
```

```python
'''2. Напишите программу, которая запрашивает у пользователя целое число и проверяет,
является ли оно числом Армстронга. Число Армстронга - это число, которое равно сумме
своих цифр, возведенных в степень, равную количеству цифр в числе.
Выведите соответствующее сообщение на экран с помощью команды print.
Пример вывода:
Введите целое число: 153
Число является числом Армстронга.'''

x2 = x1 = x =int(input('Введите целое число: '))

l = 0 
while x1 > 0:
    l = l + 1
    x1 = x1 // 10

summa = 0
while x > 0:
    summa = summa +  (x%10)**l
    x //= 10
    
if x2 == summa:
    print(f' Число является числом Армстронга. ')
else:
    print(f' Число не является числом Армстронга. ')    

```
