```python
# 2024.01.04 HW 6 
'''
1 Напишите программу, которая генерирует случайное число от 1 до 100, а затем предлагает
пользователю угадать это число. Если пользователь угадывает число, программа выводит
сообщение о победе. Если пользователь не угадывает число, программа сообщает, больше или
меньше загаданное число и предлагает попробовать снова. Используйте цикл с инструкцией
break, чтобы остановить выполнение цикла, когда число угадано.
'''
import random

x = random.randint(1,100)
#print(x)
number = int(input('Угадайте число от 1 до 100: '))

while True:        
    if x == number:
        print('Поздравляю! Вы угадали число ', number, '!' )
        break
    elif x > number:
        print('Загаданное число больше.' )
    elif x < number:
        print('Загаданное число меньше.' )  
    
    number = int(input('Попробуйте снова: '))    
```

```python
'''
2 Напишите программу, которая запрашивает у пользователя число N и выводит первые N чисел
Фибоначчи. Числа Фибоначчи - это последовательность чисел, где каждое следующее число
является суммой двух предыдущих чисел (начиная с 0 и 1). Используйте цикл while для решения
задачи. Выведите числа через запятую с помощью команды print.

'''
def Fib(n):
    if n == 0 or n == 1: return n
    else: return Fib(n-1) + Fib(n-2) 

number = int(input('Введите число N:'))
print(f'Первые {number} чисел Фибоначчи: ' , end = '')

for i in range(number):
    if i < number-1: print(f'{Fib(i)},', end = '')
    elif i == number-1: print(f'{Fib(i)}' )     
```

```python
'''
3 Напишите программу, которая запрашивает у пользователя целое положительное число и
проверяет, является ли оно простым. Простое число - это число, которое делится только на 1 и
на само себя без остатка. Используйте цикл while и проверку деления числа на все числа от 2 до
N-1 для решения задачи. Выведите соответствующее сообщение на экран с помощью команды
print.
'''

number = int(input('Введите целое положительное число:'))

d = 2
while True:
    if number%d == 0:
        print(f'Число {number} не является простым.')
        break
    else:   
        d += 1
    if d == number:
        print(f'Число {number} является простым.')
        break
```
