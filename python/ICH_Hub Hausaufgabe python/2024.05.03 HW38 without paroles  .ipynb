```python
# 2024.05.03 HW 38
''' 1 Доработать мини-интерфейс к базе данных, который был сделан на занятии. 
Новые возможности интерфейса:
1. Ввести список полей выбранной таблицы.
2. При вводе искомого значения добавить возможность выбора знака - найти записи, 
в которых выбранное поле больше, меньше или равно введенному значению.'''

import mysql.connector   

def data_columns():
    # 1 соединение
    dbconfig = {'host': 'xxxxxxxxxxxxxxxxxxxxxxxxx.amazonaws.com',
            'user': 'xxxxxxxxxx',
            'password': 'xxxxxxxxxxx',
            'database': 'xxxxxxxxxxxxxxxxxxx'}
    connection = mysql.connector.connect(**dbconfig)    

    # 2 создаем курсор
    cursor = connection.cursor()
    #cursor = cconnection()

    # 3 запрос на таблицы в базе данных 
    cursor.execute("show tables")
    tables = cursor.fetchall()   
    
    # 4 определяемся с таблицей
    columns = {"users" : ["id","name","age"],
               "product": ["pid","prod","quantity"],
               "sales": ["sid","pid","id"]}

    field = input(f"1 шаг: Выберите таблицу из {tables} : ").strip()    
    if field not in columns :  # [table]
        print(f'! Введено некорректное значение таблицы: {field} ')
        return False
    print( f'Задание № 1: Таблица {field}, поля: {columns[field]} ') 
    print( f'Данные по таблице {field} ' ) 
            
    # 5 задание 1: вывод полей таблицы
    #print( f'\t{"  ".join(columns[field])}' ) # !!!! формат вывода    
    for element in columns[field]:
        print( element, '\t', end='')
    print()

    cursor.execute(f"SELECT * FROM {field}")
    result = cursor.fetchall()
    for line in result: 
        for el in line:            
            print( f"{el:<10}",end = '' )  
        print()      
        
    # --------------
    # 6 диалог о критерии отбора информации
    print( f'Задание № 2: Отбор информации в таблице {field}: ' ) # sid
    print( f' - Выберите поле таблицы {field}, по которому отбираем инфформацию!' ) # sid
    criterium = input(f"2 шаг:\n- Выберите поле, по которому отбираем инфформацию {columns[field]} : ").strip()
    print( f' - Вы ввели поле: {criterium} ' )  # 5 
    
    print( f" - Введите числовое значение для отбора информации по выбранному полю {criterium }! " )
    xcriterium = int( input("Введите число n" ) )  
    print( f' - Вы ввели число: {xcriterium} ' )  # 5 

    print( f" - Уточните область для поиска информации {criterium } > или = или < {xcriterium}! " )
    ycriterium = input(' Введите область поиска > n: - введите "> = <" ! ') 
    print( f' - Вы ввели критерий отбора значений: {ycriterium} ' ) # 1
    print()
    
    columns = {"users" : ["id","name","age"],
               "product": ["pid","prod","quantity"],
               "sales": ["sid","pid","id"]}
    # таблица field,  поле-критерий criterium, граничное значение xcriterium, область для поиска информации ycriterium
 
    print(f' Данные по таблице {field} по критерию: {criterium} {ycriterium } {xcriterium } ') 
    for element in columns[field]:
        print( element, '\t', end='')
    print()
    
    for index  in range(len(columns[field])) :  #in 0,1,2:
        if criterium == columns[field][index]: 
            #print(f' {field} ')
            for line in result:  
                try:  
                    if int( line[index]):
                        if ycriterium == '>':
                            if int(line[index]) > xcriterium:                                 
                                for el in line: 
                                    print( f"{el:<10}",end = '' )
                                print()
                        elif ycriterium == '=':
                            if int(line[index]) == xcriterium:                                 
                                for el in line: 
                                    print( f"{el:<10}",end = '' )
                                print() 
                        elif ycriterium == '<':
                            if int(line[index]) < xcriterium:                               
                                for el in line: 
                                    print( f"{el:<10}",end = '' )
                                print()

                except ValueError:  
                    print(f' ! Выбранное поле текстового типа. С текстом сейчас программа не работает.') 
                    return False                                                         
    return True   
    
print( data_columns() ) 
```
