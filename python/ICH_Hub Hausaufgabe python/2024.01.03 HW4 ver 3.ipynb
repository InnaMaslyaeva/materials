```python
'''
2 Напишите программу, которая запрашивает у пользователя два логических значения (True или
False) и выводит результаты следующих логических операций:
● Логическое И (and) между двумя значениями.
● Логическое ИЛИ (or) между двумя значениями.
● Логическое НЕ (not) для каждого значения.
● Результат сравнения двух значений на равенство.
● Результат сравнения двух значений на неравенство.
Результаты должны быть выведены с помощью команды print.'''

x1 = bool(input('Введите первое логическое значение (True или False)'))
x2 = bool(input('Введите второе логическое значение (True или False)'))
print()
print(f'Результат логического И: {x1 and x2}')
print(f'Результат логического ИЛИ:  {x1 or x2}')
print(f'Результат логического НЕ (для первого значения): {not(x1)}')
print(f'Результат логического НЕ (для второго значения): {not(x2)}')
print(f'Результат сравнения на равенство:  {x1 == x2}')
print(f'Результат сравнения на неравенство:  {x1 != x2}')
```
