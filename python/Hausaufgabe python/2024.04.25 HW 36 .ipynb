```python
# 2024.04.25 HW 36
''' Напишите программу, которая запрашивает у пользователя 
URL-адрес веб-страницы, использует библиотеку Beautiful Soup 
для парсинга HTML и выводит список всех ссылок на странице. '''

import requests
from bs4 import BeautifulSoup

def emails_parsing(link):
    html = requests.get(link).text
    soup = BeautifulSoup(html, "html.parser")

    links = soup.find_all("a")
    for i in links:
        href = i.attrs.get("href")
        if href:
            if href[:4]=="http":
                print(href)

link = input('input url ')
#link = "https://habr.com/ru/feed/"
#link = 'https://www.mongodb.com/atlas'

emails_parsing(link)
```
